## Word cloud generation

```python
from matplotlib.pyplot import *
from wordcloud import WordCloud, STOPWORDS
from PIL import Image

def generate_word_cloud_jupyter():
    print("--- Word Cloud Generator ---")
    user_text=input("Enter your text: ")

    stopwords = STOPWORDS

    wordcloud_params = {
        'width': 1000,
        'height': 600,
        'background_color': 'black',
        'stopwords': stopwords,
        'min_font_size': 10,
        'max_font_size': 180,
        'max_words': 250,
        'colormap': 'viridis',
        'random_state': 42,
        'collocations': False,
        'prefer_horizontal': 0.9
    }

    print("\nGenerating word cloud...")
    wordcloud = WordCloud(**wordcloud_params).generate(user_text)

    figure(figsize=(15, 10))
    imshow(wordcloud, interpolation='bilinear')
    axis('off')
    tight_layout(pad=0)
    show()

    print("\nWord cloud displayed successfully!")
```

```python
# Cell 3: Run the word cloud generator
generate_word_cloud_jupyter()
```
